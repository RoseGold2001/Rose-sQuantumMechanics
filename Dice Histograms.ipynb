```python
import matplotlib.pyplot as plt
import random

D6_1 = random.randint(1, 7)
D6_2 = random.randint(1, 7)


print(f"The D6 rolled: {D6_1}")
print(f"The D6 rolled: {D6_2}")

Roll= D6_1 + D6_2
print(Roll)

#Got a one time roll

#I DESPISE iterating but I can't think of another way to do so.
Rolls = {}
print('The possible values for rolls are as follows:')
TotalRoll = 0
for D6_1 in range(1, 7):  #Idk why it doesn't count the last number 
    for D6_2 in range(1, 7): 
        Roll = D6_1 + D6_2
        print(f"D1= {D6_1}, D2= {D6_2}, Total: {Roll}")
        if Roll in Rolls:
            Rolls[Roll] += 1
        else:
            Rolls[Roll] = 1
        TotalRoll+=1
        print(TotalRoll)

# this makes sense but again, people online do it. Idk how i'd have figured it out on my own.
rolls_values = list(Rolls.keys())
occurrences = list(Rolls.values())
probabilities = [(count / TotalRoll) * 100 for count in occurrences]

print('The possible values for rolls and their probabilities are as follows:')
for roll, count in Rolls.items():
    probability = (count / TotalRoll)* 100, 'Percent' # There are 36 total possible combinations (6 * 6)
    print(f"Roll: {roll}, Probability: {probability}")
    

plt.bar(rolls_values, probabilities, color='purple')
plt.xlabel('Roll Amount D6 + D6')
plt.ylabel('Chance')
plt.title('Dice Rolls')
plt.show()

```

```python
import matplotlib.pyplot as plt
import random

# Simulate a coin flip (1 for heads, 2 for tails)
flip_result = random.randint(1, 2)

if flip_result == 1:
    print("Flip result: Heads")
else:
    print("Flip result: Tails")

# Running simulation
heads_count = 0
tails_count = 0
total_flips = 1000000 # to the total number of flips, can be increased

for _ in range(total_flips):
    if random.randint(1, 2) == 1:
        heads_count += 1
    else:
        tails_count += 1

# Calculate probabilities as percentages
heads_probability = (heads_count / total_flips) * 100
tails_probability = (tails_count / total_flips) * 100

print(f"Heads Probability: {heads_probability}%")
print(f"Tails Probability: {tails_probability}%")

# Plotting the bar chart
flip_labels = ['Heads', 'Tails']
flip_probabilities = [heads_probability, tails_probability]

plt.bar(flip_labels, flip_probabilities, color=['purple', 'green'])
plt.xlabel('Coin Flip Outcome')
plt.ylabel('Probability')
plt.title('Coin Flip Probability')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

Rolls = {}
TotalRoll = 0
count_twenty = 0

for D20_1 in range(1, 21):
    for D20_2 in range(1, 21):
        Roll = max(D20_1, D20_2)  
        if Roll == 20:
            count_twenty += 1 
        TotalRoll += 1
        if Roll in Rolls:
            Rolls[Roll] += 1
        else:
            Rolls[Roll] = 1


probability_twenty = (count_twenty / TotalRoll) * 100

print('The probability of getting a 20 as the largest roll:', probability_twenty)


```
